```python
from pymongo import MongoClient
from datetime import datetime

# Connexion à MongoDB
client = MongoClient("mongodb://admin:admin@localhost:27017/")
db = client["Bank_DB_client"]
collection = db["Bank_client"]

# Migration des dates
for doc in collection.find({"Date": {"$type": "string"}}):  # Cibler uniquement les dates en string
    try:
        date_str = doc["Date"]
        date_obj = datetime.strptime(date_str, "%Y-%m-%d")  # Convertir la string en objet datetime
        collection.update_one(
            {"_id": doc["_id"]},
            {"$set": {"Date": date_obj}}  # Mettre à jour avec un objet Date
        )
        print(f"Document {doc['_id']} mis à jour ✅")
    except Exception as e:
        print(f"Erreur sur le document {doc['_id']} : {e}")
```
